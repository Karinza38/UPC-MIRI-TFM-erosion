```python
import networkx as nx
import numpy as np

# Create a weighted graph
G = nx.Graph()

G.add_edge(0, 1, weight=0.8)
G.add_edge(1, 2, weight=0.5)
G.add_edge(0, 2, weight=0.7)
G.add_edge(0, 0, weight=0.1)

# IO: https://networkx.org/documentation/stable/reference/readwrite/matrix_market.html

# Convert the graph to a weighted adjacency matrix
adj_matrix = nx.to_numpy_array(G, weight="weight")
print("adj_matrix", type(adj_matrix), ":\n", adj_matrix)

# Convert the matrix to a NumPy array
adj_matrix_array = np.array(adj_matrix)
print("adj_matrix_array", type(adj_matrix_array), ":\n", adj_matrix_array)

# Draw it
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_size=1000, node_color="skyblue", font_size=12, font_color="black", font_weight="bold")

# Create a dictionary of edge weights
edge_labels = {(i, j): f"{data['weight']:.2f}" for i, j, data in G.edges(data=True)}
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_size=10, font_color="red")
nx.draw

plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a weighted directed graph
G = nx.DiGraph()

G.add_edge(1, 2, weight=0.8)
G.add_edge(2, 3, weight=0.5)
G.add_edge(3, 1, weight=0.7)
G.add_edge(3, 2, weight=0.9)
G.add_edge(1, 1, weight=0.1)

# Draw the directed graph with edge weights and labels
#pos = nx.spring_layout(G)
pos = nx.circular_layout(G)
# Adjust edge positions to avoid overlapping nodes
pos_off = pos.copy()
for node, (x, y) in pos.items():
    pos_off[node] = (x, y + 0.0)  # Adjust y-coordinate slightly

nx.draw(G, pos_off, with_labels=True, node_size=1000, node_color="skyblue", font_size=12, font_color="black", font_weight="bold")

# Adjust edge positions to avoid overlapping nodes
node_pos = pos.copy()
for node, (x, y) in pos.items():
    node_pos[node] = (x, y + 0.5)  # Adjust y-coordinate slightly

# Create a dictionary of edge labels
edge_labels = {}
for i, j, data in G.edges(data=True):
    label = f"{data['weight']:.2f}"
    edge_labels[(i, j)] = label


nx.draw_networkx_edge_labels(G, node_pos, edge_labels=edge_labels, font_size=10, font_color="red")

# Display the plot
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a weighted directed graph
G = nx.DiGraph()

G.add_edge(1, 2, weight=0.8)
G.add_edge(2, 3, weight=0.5)
G.add_edge(3, 1, weight=0.7)
G.add_edge(3, 2, weight=0.9)

# Create a dictionary of edge labels
edge_labels = {(i, j): f"{data['weight']:.2f}" for i, j, data in G.edges(data=True)}

# Draw the directed graph with edge weights and labels
pos = nx.circular_layout(G)  # Circular layout

# Create the figure and axis
fig, ax = plt.subplots()

# Draw edges with adjusted z-order
nx.draw_networkx_edges(G, pos, ax=ax, edge_color="black", width=1.0, arrowstyle="->", arrowsize=10)

# Draw nodes with adjusted z-order
nx.draw_networkx_nodes(G, pos, ax=ax, node_size=1000, node_color="skyblue")

# Draw labels with adjusted z-order
nx.draw_networkx_labels(G, pos, ax=ax, font_size=12, font_color="black", font_weight="bold")

# Draw edge labels with adjusted z-order
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_size=10, font_color="red")

# Set z-order of the plot
ax.set_zorder(1)

# Display the plot
plt.show()
```

```python
# percolation
# https://stackoverflow.com/questions/54995434/trying-to-use-percolation-centrality-from-networkx
# https://networkx.org/documentation/stable/reference/algorithms/generated/networkx.algorithms.centrality.percolation_centrality.html

import networkx as nx
import matplotlib.pyplot as plt
import random

# Parameters
grid_size = 5  # Size of the grid
p = 0.5  # Probability of occupying a node

# Create a graph (grid)
G = nx.grid_2d_graph(grid_size, grid_size)
print(f"G [{len(G)}] :\n", G)

# Occupy nodes with probability p
for node in G.nodes:
    if random.random() <= p:
        G.nodes[node]['occupied'] = True
    else:
        G.nodes[node]['occupied'] = False

# Create subgraphs of occupied nodes
occupied_subgraphs = list(nx.connected_components(G.subgraph((node for node in G.nodes if G.nodes[node]['occupied']))))
print(f"occupied_subgraphs [{len(occupied_subgraphs)}] :\n", occupied_subgraphs)

# Plot the occupied clusters
pos = {(x, y): (x, -y) for x, y in G.nodes}
for i, cluster in enumerate(occupied_subgraphs):
    nx.draw(G.subgraph(cluster), pos, with_labels=True, node_size=500, label=f"Cluster {i + 1}")

# analysis
perc = nx.percolation_centrality(G, "occupied")
print(f"perc [{len(perc)}] :\n", perc)

for i, cluster in enumerate(occupied_subgraphs):
    perc = nx.percolation_centrality(G.subgraph(cluster), "occupied")
    print(f"perc_{i} [{len(perc)}] :\n", perc)

# graph
plt.title(f"Site Percolation (p = {p})")
plt.legend()
plt.show()
```
