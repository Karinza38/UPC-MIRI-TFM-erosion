```python
import numpy as np
import matplotlib.pyplot as plt

def plot2D(f, res):
    sizeX = 20
    sizeZ = 10
    resX = int(res*sizeX)
    resZ = int(res*sizeZ)

    # translate to mimic scene
    x = np.linspace(-10.5, 9.5, resX)
    y = np.linspace(-1, 9, resZ)

    # Create a grid of (x, y) points
    X, Y = np.meshgrid(x, y)

    # Calculate the heat values for each (x, y) point in the grid
    Z = f(X, Y)
    print(Z.min(), Z.max())

    ## Normalize the heat values to the range [0, 1]
    #from matplotlib.colors import Normalize
    #norm = Normalize(vmin=Z.min(), vmax=Z.max())
    #Z = norm(Z)

    # Create the heatmap
    plt.imshow(Z, extent=[x.min(), x.max(), y.min(), y.max()], origin='lower', cmap='coolwarm', aspect='equal')

    # Add labels and a colorbar
    plt.xlabel('X')
    plt.ylabel('Z')
    plt.colorbar(label='Normalized R')

    # Show the plot
    plt.title('Resistance field 2D')
    plt.show()
```

```python
def pockets(X, Y):
    R = np.sin(X) + np.cos(Y)
    R = (R+2.0) / 4.0 # normalize
    return R

plot2D(pockets, 8)
```

```python
#https://www.wolframalpha.com/input?i=plot+%280.5*sin%28%2810*x%2B5*y%29%29%2B0.5%29+*+step+function%28sin%2820*y%29+%2B+0.8%29++for+x%3D-1%2C1+y%3D-1%2C1

def wavy(X, Y):
    R = np.sin(-1 * X + 0.5 * Y)
    R = (0.5 * R + 0.5) # normalize
    return R

plot2D(wavy, 8)
```
